# Series
1D Labeled data that can hold any data type

```python
data = [100 , 102, 104]

series = pd.Series(data, index=["a", "b", "c"])

print(series)

print(series.loc["a"]) # Location by Label

series.loc["c"]=200

print(series)

print(series.iloc[1]) # Integer Location/ Location by ID
```

```python
data = [100 , 102, 104, 200, 202]

series = pd.Series(data, index=["a", "b", "c", "d", "e"])

print(series[series>=200])

print(series[series<200])
```

```python
# Dictionary

calories = {"Day 1":1750, "Day 2": 2100, "Day 3":1700}

series= pd.Series(calories)
print(series)

print(series.loc["Day 2"])


print("Calories More than 2000: ", series[series>=2000])

print("Calories less than 2000: ", series[series<2000])

series.loc["Day 3"]+=500

print(series)
```
