# Data Cleaning
Process of fixing/removing :
        incomplete, incorrect, irrelevent data
        ~75% work done by pandas is Data Cleaning  

```python
import pandas as pd
df = pd.read_csv("data.csv")

print(df)
```

### Drop irrelevent Data

```python


df = df.drop(columns=["Legendary", "No"])

print(df)
```

### Handle Missing Data

```python
# df = df.dropna(subset=["Type2"])
# print(df.to_string())

df = pd.read_csv("data.csv")

df = df.fillna({"Type2": "None"})
print(df.to_string())
```

### Fix inconsistent values

```python
df = pd.read_csv("data.csv")

df["Type1"] = df["Type1"].replace(
                                    {
                                        "Grass": "GRASS", 
                                        "Fire":"FIRE", 
                                        "Water":"WATER"
                                    }
                                    )

print(df.to_string()) 
```

### Standardize Text

```python
df = pd.read_csv("data.csv")

df["Name"] = df["Name"].str.lower()
print(df.to_string()) 

```

### Fix Data Type

```python
df = pd.read_csv("data.csv")

df["Legendary"] = df["Legendary"].astype(bool)
print(df.to_string())
```

### Remove Duplicate Values

```python
df = pd.read_csv("data.csv")
print(df)
df = df.drop_duplicates()
print(df.to_string())
```
