# Importing

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")

print(df.to_string())
```

# Selection

```python
#Selection by Column

# print(df["Name"].to_string())

# print(df["Height"].to_string())

# print(df["Weight"].to_string())

print(df[["Name", "Height", "Weight"]].to_string())


```

```python
# Selection By Row 

# print(df.loc[1])

df=pd.read_csv("data.csv", index_col="Name") # Made Name as index column  
print(df.loc["Charizard"])    # Accessed Data using Name_Column as index 
print(df.loc["Charizard", ["Height", "Weight"]]) # Only mentioned things will get displayed
print(df.loc["Charizard":"Blastoise", ["Height", "Weight"]]) #Print data from Charizard to Blastoise
```

```python
# Selection by Integer Indexing

print(df.iloc[0:11])

print(df.iloc[0:11:2])

print(df.iloc[0:11, 0:3]) #Print Data from 0th row to 11th upto 3 columns
```
