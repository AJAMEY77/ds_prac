# DataFrames 
Tabular Data Structure with rows and column(2D) Similar to excel 

```python
import pandas as pd
```

```python
data={
    "Name": ["SpongeBob", "Patrick", "Squidward"],
    "age": [30, 35, 50] 
    }

df = pd.DataFrame(data, index=["Employee 1", "Employee 2", "Employee 3"])

print(df)

print(df.loc["Employee 1"]) # Access by Labeled Index

print(df.iloc[2])  # Access by Integer Index 

# Add a new Column 
df["Job"]=["Cook", "Intern", "Cashier"]

print(df)


# Add a new Row
new_Row = pd.DataFrame([
    {"Name": "Sandy", "age":28, "Job": "Engineer"},
    {"Name":"Mr Crab", "age":60, "Job":"Manager"}
],
                       index=["Employee 4", "Employee 5"]) 

df = pd.concat([df, new_Row])

print(df)
```
