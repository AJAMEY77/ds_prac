## 1D Array

```python
import numpy as np

print(np.__version__)
```

```python
my_list=[1,2,3,4]
print(my_list)

my_list=my_list*2
print(my_list)
```

```python
array= np.array([1,2,3,4])
print(array)

print(type(array)) #ndarray: N Dimensional Array


array=array*2
print(array)
```

## MultiDimensional Array

```python
array = np.array([
    [
    ['A', 'B', 'C'], 
    ['D', 'E', 'F'], 
    ['G', 'H', 'I']
    ],
    [
    ['A', 'B', 'C'], 
    ['D', 'E', 'F'], 
    ['G', 'H', 'I']
    ],
    [
    ['A', 'B', 'C'], 
    ['D', 'E', 'F'], 
    ['G', 'H', 'I']
    ]
    ])
print(array.ndim) #number of dimension of this 'array'
print(array.shape) #Depth Num_of_rows Num_of_col
 
print(array[0][0][0]) #Chain indexing 
print(array[0,0,0])   #MultiDimensional Indexing
```
