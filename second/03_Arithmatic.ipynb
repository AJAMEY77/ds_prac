## Scaler Arithmatic

```python
import numpy as np

# Scaler Arithmatic : Single Value
array = np.array([1,2,3])
print(array)

print(array+1)
print(array-2)
print(array*3)
print(array/4)
print(array**5)


```

## Vectorized Math Function 


```python
import numpy as np

# Vector : Linear Algebra term/ One D Array 
array = np.array([1.01, 2.5, 3.99])

print(np.sqrt(array))
print(np.round(array))
print(np.floor(array))
print(np.ceil(array))

#Built in constant
print(np.pi)




```

```python
radii = np.array([1,2,3])

print(np.pi*radii ** 2)
```

# Element Wise Arithmatic

```python
array1=np.array([1,2,3])
array2=np.array([4,5,6])

print(array1+array2)
print(array1-array2)
print(array1*array2)
print(array1/array2)
print(array1**array2)




```

# Comparison Operator

```python
scores =np.array([91, 55, 100, 73, 82, 64])
print(scores== 100)
print(scores>60)
print(scores<60)

scores[scores<60]=00
print(scores)
```
