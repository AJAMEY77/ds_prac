# Random Number

```python
import numpy as np
```

```python
rng = np.random.default_rng(seed=1) #seed will return same result
print(rng.integers(low=1, high=101, size=(3,2)))
```

```python
np.random.seed(1)
print(np.random.uniform(low=-1, high=1, size=(3, 2))) #Uniform Distribution 
```

```python
rng = np.random.default_rng()

array= np.array([1,2,3,4,5])

rng.shuffle(array)

print(array) 
```

```python
rng = np.random.default_rng()

fruits = np.array(["🍎", "🍊", "🍌", "🥥","🍍"]) 
fruit = rng.choice(fruits, size=(3,3))
print(fruit)
```
