## Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dat Structure : [restaurant_id, 2021, 2022, 2023, 2024]
sales_data=np.array([
    [1, 150000, 180000, 220000, 150000],
    [2, 120000, 140000, 160000, 190000],
    [3, 200000, 230000, 260000, 300000],
    [4, 180000, 210000, 240000, 270000],
    [5, 160000, 185000, 205000, 230000]
])

print("=== Zomato Sales Analysis ===")
print("\n Sales data shape : ", sales_data.shape)
print("\n Sample Data for 1st 3 restaurant :\n", sales_data[:3])
```

```python
#Total Sales Per Year 
print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:,1:], axis=0)
print(yearly_total)
```

```python
# Minimum Sales per restauarant
min_sales=np.min(sales_data[:, 1:], axis=1)
print("Min Sales Per Restaurant : ", min_sales)
```

```python
#Maximum sales per year 
max_sales = np.max(sales_data[:,1:], axis=0)
print(max_sales)

#Average sales per restaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sale : ", avg_sales)
```

```python
#Cumulitive Sales 
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulitive Sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show() 
```
